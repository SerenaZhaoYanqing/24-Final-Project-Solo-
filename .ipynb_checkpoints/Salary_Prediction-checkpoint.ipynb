```python
#Dependencies
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
```


# Load Data

```python
#read csv file 
df=pd.read_csv('survey_results_public.csv')
df.head()
```

# Perform EDA 

```python
#clean data--Keep revevant columns 
#EdLevel: Education level
#YearsCodePro: Years of professional experience 
#ConvertedCompYearly: converted yearly salary in USD 
df = df[["Country", "EdLevel","LearnCode", "YearsCodePro", "Employment", "ConvertedCompYearly","Age","Gender"]]
df = df.rename({"ConvertedCompYearly": "Salary"},axis=1)
```

```python
#drop rows where cell contains null value
df=df.dropna()
#should contain the same rows for all columns 
df.info()
```

```python
#Countries

#clean countries where we only have a few values 
counts=df['Country'].value_counts()
# using quantile, finding over 90% of the countries has more than about 413 rows data
counts.quantile([.9])

#drop countries where there are less than 413 rows of data
to_remove =counts[counts <= 413].index

# Keep rows where the country value counts is not in to_remove
df = df[~df.Country.isin(to_remove)]
```

```python
df['Country'].unique().tolist()
```

```python
#Edlevel:Edcuation Level

#check details of educational level 
df["EdLevel"].unique()

#standarlise the educational level 
def educational_level(x):
    if 'Bachelor’s degree' in x:
        return 'Bachelor’s degree'
    if 'Master’s degree' in x:
        return 'Master’s degree'
    if 'Professional degree' in x:
        return 'Professional degree'
    if 'Other doctoral degree' in x:
        return 'PHD' 
    return 'Less than a Bachelors'

df['EdLevel'] = df['EdLevel'].apply(educational_level)
df["EdLevel"].unique()
```

```python
df['EdLevel'].unique().tolist()
```

```python
df['Age'].unique().tolist()
```

```python
#LearnCode: How to learn coding/Where 

#Check details of the survey 
df['LearnCode'].unique()

#standarlise the learn code channel 
def learn_code(x):
    if 'online' in x:
        return 'Online Resources'
    if 'Bootcamp' in x:
        return 'Bootcamp'
    if 'School' in x:
        return 'School'
    return 'Others'

df['LearnCode'] = df['LearnCode'].apply(learn_code)
df['LearnCode'].unique()
```

```python
# Yearscodepro: Professional Year of Experience 

#check details of professional experience (in years)
df["YearsCodePro"].unique()

#define a function to replace "less than 1 year" & greater than 50 years
def professional_experience(x):
    if x ==  'More than 50 years':
        return 51
    if x == 'Less than 1 year':
        return 0.5
    return float(x)

df['YearsCodePro'] = df['YearsCodePro'].apply(professional_experience)
```

```python
#Employment
df['Employment'].unique()

#standarlise the employment and remove the retired and results that didnt include employment status 
def employment_status(x):
    if 'full-time' in x:
        return 'Full Time'
    if 'part-time' in x:
        return 'Part Time'
    if 'self-employed' in x:
        return 'Self-Employed'
    return 'Others'

df['Employment'] = df['Employment'].apply(employment_status)
df = df[df['Employment']!='Others']

```

```python
#Salary

#getting quantile value of 75% and 25%, getting in between.
salary = df["Salary"]
salary.quantile([.25])
salary.quantile([.75])

# keep the range of salary
df = df[df["Salary"] <= 109672]
df = df[df["Salary"] >= 41520]
df.head()
```

```python
#Gender 
df["Gender"].unique()
def gender_stardarlised(x):
    if 'Woman' in x:
        return 'Female'
    if 'Man' in x:
        return 'Male'
    return 'Others'

df['Gender'] = df['Gender'].apply(gender_stardarlised)
df = df[df['Gender']!='Others']
df.head()
```

# Are salaries normally distributed? 

```python
# check the kurtosis and skewness of salary

print('Salary Skewness:', df['Salary'].skew())
print('Salary Kurtosis:', df['Salary'].kurt())
print('\n')
if -0.5 <= df['Salary'].skew() <= 0.5:
    print('Salary distribution is approximately symmetric')
elif -0.5 >  df['Salary'] > 0.5:
    print('Salary distribution is skewed')
    
```

# Visualization --Categorical data with Salary 

Next we visualize each categorical feature to see which features could be good predictors of salary. We use the violin plot here to see the shape of the distribution curve along with mean and interquartile ranges.

```python
def rel_cat(df,col):
    #Make subplots - left be the distribution of samples on the feature and right be the dependance of salary on the feature
    plt.figure(figsize=(12,8))
    plt.subplot(1,2,1)
    if df[col].dtype == 'int64':
        df[col].value_counts().sort_index().plot()
    else:
        #else change the categorical variable to category type and order their level by mean salary 
        mean=df.groupby(col)['Salary'].mean()
        df[col]=df[col].astype('category').copy()
        levels = mean.sort_values().index.to_list()
        df[col].cat.reorder_categories(levels, inplace = True)
        df[col].value_counts().plot()
    plt.xticks(rotation = 40)
    plt.xlabel(col)
    plt.ylabel('Counts')
    
    plt.subplot(1,2,2)
    if col == 'Country': 
        sns.boxplot(x=col, y = df['Salary'], data = df)
        plt.xticks(rotation = 90)
    elif df[col].dtype == 'category':
        sns.violinplot(x = col, y = df['Salary'], data = df, scale = 'count', inner = 'quartile')
        plt.xticks(rotation = 70)
    plt.ylabel('Salaries')
    
```

```python

rel_cat(df, 'Country')
plt.savefig('image/country_salarymean.png')
plt.show()
```

# Working on here 

```python
#target data : salary 
target=df["Salary"]
train_target_df=target.to_frame()
train_target_df.head()
```

```python
#testing feature data, removed salary  (it will be the same as training features data)

test_feat_df=df[['Country','EdLevel','LearnCode','Age','YearsCodePro']]
train_feat_df=test_feat_df.copy()
```

```python
#define catorical & numerical &target variable
cat_vars=['Country','EdLevel','LearnCode','Age']
num_vars=['YearsCodePro']
tar='Salary'
```

```python
print(train_feat_df.shape)
print(train_target_df.shape)
print(test_feat_df.shape)
```

```python
#full training data (full set)
train_df=df.copy()
train_df.head()
print(train_df.shape)
```

```python
print(train_df.dtypes)
```

```python
#one hot encode categorical data in train_df dataset
train_df=pd.get_dummies(train_df)
```

```python
train_df.info()
```

```python

#Split the data and assign 'salary' to 'sal_df' and the rest of the features to 'feat_df'. 
X_df = train_df[train_df.loc[ :, train_df.columns != 'Salary'].columns]
y_df = train_df['Salary']
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size = 0.2, random_state = 1)
```

# linear Regression

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
lr
```

```python
#view the coefficients of our model
print(lr.intercept_)
print(lr.coef_)
```

```python
#predict on training data
y_pred_lr=lr.predict(X_train)

#score the prediction with mse and r2 
mse_lr=mean_squared_error(y_train,y_pred_lr)
r2_lr=r2_score(y_train,y_pred_lr)
score_lr=cross_val_score(lr,X_train,y_train)
print(f"Mean Squared Error (MSE): {mse_lr}")
print(f"R-squared (R2 ): {r2_lr}")
print(f"Score: {score_lr}")
```

```python
#plot the actual vs predict

def dis_data(Rfunction, Bfunction, Rname, Bname, title):
    plt.figure(figsize=(10,6))
    ax1 = sns.distplot(Rfunction, hist = False, color = 'r', label = Rname)
    ax1 = sns.distplot(Bfunction, hist = False, color = 'b', label = Bname)
    plt.title(title)

dis_data(y_train, y_pred_lr,'Actual Salary (train)','Predicted Salary (train)', 
         title='Predicted VS Actual Salary(train data)')
plt.savefig('image/linearregression.png')
```

##### We will do the same for the rest 20% test data to see if result is similar


```python
y_pred_test_lr=lr.predict(X_test)
mse_lr_test=mean_squared_error(y_test,y_pred_test_lr)
r2_lr_test=r2_score(y_test,y_pred_test_lr)
score_lr_test=cross_val_score(lr,X_test,y_test)
print(f"Mean Squared Error (MSE): {mse_lr_test}")
print(f"R-squared (R2 ): {r2_lr_test}")
print(f"Score: {score_lr_test}")
```

###### It is showing similar result, we will need to improve it 

## Ridge Regression

```python
#create a ridge regression object and fit it to training data 
RidgeModel = Ridge(alpha = 1.0)
RidgeModel.fit(X_train,y_train)
```

```python
#predict values of training data and train data
y_pred_Ridge_train = RidgeModel.predict(X_train)
#R-squared & MSE  of training and testing data - 
print ('R-squared values(train)', RidgeModel.score(X_train, y_train))
print ('MSE of training data', mean_squared_error(y_train, y_pred_Ridge_train))
print(f'Score:', cross_val_score(RidgeModel,X_train,y_train))
```

```python
#predict values of training data and train data
y_pred_Ridge_test = RidgeModel.predict(X_test)
#R-squared & MSE  of training and testing data - 
print ('R-squared values(test)', RidgeModel.score(X_test, y_test))
print ('MSE of training data', mean_squared_error(y_test, y_pred_Ridge_test))
print(f'Score:', cross_val_score(RidgeModel,X_test,y_test))
```

# Grid Search
